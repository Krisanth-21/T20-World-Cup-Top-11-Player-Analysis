```python
import pandas as pd
import json
```

```python
file_path = 't20_json_files/t20_json_files/t20_wc_match_results.json'

with open(file_path) as f:
    file_path = json.load(f)
df_match = pd.DataFrame(file_path[0]['matchSummary'])

df_match
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match
```

```python
match_ids_dictionary = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']

    match_ids_dictionary[key1] = row['match_id']
    match_ids_dictionary[key2] = row['match_id']
match_ids_dictionary
```

```python
df_match.to_csv('t20_csv_files/dim_match_summary.csv', index = False)
```

<h3> Batting Summary</h3>


```python
file_path = 't20_json_files/t20_json_files/t20_wc_batting_summary.json'

with open (file_path) as f:
    data = json.load(f)

all_records = []
for record in data:
    all_records.extend(record['battingSummary'])
                      
df_batting = pd.DataFrame(all_records)
df_batting.head(20)
```

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
df_batting.head(20)
```

```python
df_batting 
```

```python
df_batting.drop(columns=['dismissal'], inplace = True)
df_batting.head(10)
```

```python
# removing special characters in batsmanName column
if 'batsmanName' in df_batting.columns:
        df_batting['batsmanName'] = df_batting['batsmanName'].str.replace('[()/x]', '', regex = True)
df_batting.head(30)
#df_batting.iloc[80:120]
```

```python
match_ids_dictionary['Namibia Vs Sri Lanka']
 
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dictionary)
df_batting.head()
```

```python
df_batting.to_csv('t20_csv_files/fact_batting_summary.csv', index = False)
```

<h3>Bowling Summary</h3>

```python
file_path = 't20_json_files/t20_json_files/t20_wc_bowling_summary.json'

with open (file_path) as f:
    data = json.load(f)

all_records = []
for record in data:
    all_records.extend(record['bowlingSummary'])
                      
all_records[:2]
```

```python
df_bowling = pd.DataFrame(all_records)
df_bowling.shape
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids_dictionary)
df_bowling.head()
```

```python
df_bowling.to_csv('t20_csv_files/fact_bowling_summary.csv', index = False)
```

<h4> Players Information </h4>

```python
file_path = 't20_json_files/t20_json_files/t20_wc_player_info.json'

with open (file_path) as f:
    data = json.load(f)
```

```python
df_players = pd.DataFrame(data)
df_players.shape
df_players.head(11)
```
